```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('2024-10-07-16-41-49/_0/n_features=16_n_clusters=1_class_sep=0.5_balance=0.1/0/k-2/u-2/fraudulent/reward-label/file.csv')

# Define the window size for rolling calculation
window_size = 100

# Calculate rolling mean for each column (assuming the first column is not needed for plotting)
rolling_df = df.iloc[:, 1:].rolling(window=window_size).mean()

# Plot the rolling mean
plt.figure(figsize=(10, 6))
plt.plot(rolling_df)
plt.title(f'Rolling Mean (Window Size = {window_size})')
plt.xlabel('Index')
plt.ylabel('Rolling Mean Value')
plt.legend(df.columns[1:], loc='upper left')
plt.show()
```
