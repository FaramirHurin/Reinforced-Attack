```python
import pandas as pd
import matplotlib.pyplot as plt

# 
# Load the CSV file
file_path = 'averaged_results/n_features=32_n_clusters=8_class_sep=1_balance=0.5_k-4_u-6_fraudulent_reward-label.csv'
data = pd.read_csv(file_path)

# Set the moving average window size (adjust as needed)
window_size = 200
moving_average = data.rolling(window=window_size).mean()

# Plot each column's moving average with its name as the label
plt.figure(figsize=(10, 6))

# Loop through each column and plot its moving average
for column in moving_average.columns:
    plt.plot(moving_average[column], label=column)

# Add labels and legend
plt.xlabel('Index')
plt.ylabel('Values')
plt.title(f'Moving Average with Window Size {window_size}')
plt.legend()
plt.show()
```
