```python
import pickle
import os
import pandas as pd

big_df = None

# Directory containing the .pkl files
directory = 'data/simulated-data-transformed/'
total_shape = 0
counter = 0
# Iterate over all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.pkl'):
        file_path = os.path.join(directory, filename)
        
        # Open and load each .pkl file
        with open(file_path, 'rb') as file:
            data = pickle.load(file)
            if big_df is None:
                big_df = data
            else:
                big_df = pd.concat([big_df, data],ignore_index=True, axis=0)
            counter += 1
            total_shape += data.shape[0]
            
print(counter, total_shape)
```
